### Fill Numerical Column

```python
num_cols = df.select_dtypes(include=np.number).columns
```

```python
num_cols
```

```python
from sklearn.impute import SimpleImputer

num_imputer = SimpleImputer(strategy='median')

df[num_cols] = num_imputer.fit_transform(df[num_cols])
```

### Fill Catagorical Column

```python
cat_cols = df.select_dtypes(include='object').columns
```

```python
cat_cols
```

```python
cat_imputer = SimpleImputer(strategy='most_frequent')

df[cat_cols] = cat_imputer.fit_transform(df[cat_cols])
```

## Check Outlier

```python
for col in num_cols:
    plt.figure(figsize=(6,3))
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot of {col}')
    plt.show()
```

```python
for col in num_cols:

    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)

    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    outliers = df[(df[col] < lower) | (df[col] > upper)]

    print(f"{col}: {len(outliers)} outliers")
```

### Handle Outlier Using IQR

```python
exclude_cols = ['customer_id','risk_status']

outlier_cols = [
    col for col in num_cols
    if col not in exclude_cols
]
```

```python

for cols in outlier_cols:

    Q1 = df[cols].quantile(0.25)
    Q3 = df[cols].quantile(0.75)

    IQR = Q3 - Q1

    Lower = Q1 - (1.5 * IQR)
    Upper = Q3 + (1.5 * IQR)

    df[cols] = df[cols].clip(Lower, Upper)

print("Outliers Treated Successfully")
```

```python
for col in outlier_cols:
    plt.figure(figsize=(6,3))
    sns.boxplot(x=df[col])
    plt.title(f'After Z_Score - {col}')
    plt.show()
```

## Scalling Data

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
```

```python
scale_cols = [
    col for col in outlier_cols
]

df[scale_cols] = scaler.fit_transform(
    df[scale_cols]
)

print("\nScaling Completed")
```
